```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
# Загрузка данных
df = pd.read_csv("/content/students.csv", delimiter=",")
df_test = pd.read_csv("/content/students_test.csv", delimiter=",")

# Создание df_cut
df_cut = df[['Age', 'Year of birth', 'Month of birthday']]

# Обучение модели линейной регрессии
linearRegression = LinearRegression()
linearRegression.fit(df_cut[['Year of birth', 'Month of birthday']], df_cut['Age'])

# Предсказание для тестовой выборки
df_test['Predicted'] = linearRegression.predict(df_test[['Year of birth', 'Month of birthday']])

# Получение предсказанного возраста для первого студента из тестовой выборки
predicted_age_first_student = df_test['Predicted'].iloc[0]
print(predicted_age_first_student)

mean_absolute_error(df_test["Predicted"], df_test["Age"])
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# Шаг 1: Загрузка данных
df_train = pd.read_csv('students.csv', delimiter=',')
df_test = pd.read_csv('students_test.csv', delimiter=',')

# Шаг 2: Подготовка данных
df_train_cut = df_train[['Age', 'Year of birth', 'Month of birthday']].dropna()
df_test_cut = df_test[['Age', 'Year of birth', 'Month of birthday']].dropna()

# Шаг 3: Обучение модели
model = DecisionTreeRegressor(max_depth=3)
model.fit(df_train_cut[['Year of birth', 'Month of birthday']], df_train_cut['Age'])

# Шаг 4: Предсказание
df_test_cut['Predicted'] = model.predict(df_test_cut[['Year of birth', 'Month of birthday']])

# Шаг 5: Оценка модели
mae = mean_absolute_error(df_test_cut['Age'], df_test_cut['Predicted'])
print("Средняя абсолютная ошибка (MAE):", mae)
```

```python
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# Шаг 1: Загрузка данных
df = pd.read_csv('/content/students.csv', delimiter=',')
df_test = pd.read_csv('/content/students_test.csv', delimiter=',')

df_cut=df[['Growth','Weight','Sex']]

df_cut=df_cut.dropna()

model = RandomForestClassifier(max_depth=2, random_state=0)

model.fit(df_cut[['Growth','Weight']].values.reshape(-1,2), y=df_cut['Sex'].values)


df_test_cut=df_test[['Growth','Weight','Sex']]
df_test_cut=df_test_cut.dropna()

result=model.predict_proba(df_test_cut[['Growth','Weight']].values.reshape(-1,2))

df_test_cut['pr class 0']=result[:,0]
df_test_cut['pr class 1']=result[:,1]
df_test_cut.head()


sorted_df = df_test_cut.sort_values(by='pr class 0', ascending=False)

sorted_df.head()
#print(sorted_df)
```
